##### 1.❓Are there better or worse days/times in terms of sales? Plot a chart to show the result

```python
coffee_df.head(1)
```

```python
# Extracting dates from transaction date
coffee_df['transaction_date'].dt.date
coffee_df['transaction_time'] = pd.to_datetime(coffee_df['transaction_time']).dt.time
```

```python
# Creating time categories function to return Morning, Noon, Evenging, Night based on transaction_time
def time_cetegories(transaction_time):
    hour  = transaction_time.hour
    if 5<= hour <12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Noon'
    elif 17<= hour < 19:
        return 'Evening'
    else:
        return 'Night'

# use apply to apply the time category into a new column called time_category
coffee_df['time_category'] = coffee_df['transaction_time'].apply(time_cetegories)
coffee_df.head(3)
```

```python
# Are there better or worse days/times in terms of sales?
by_day = coffee_df.groupby('transaction_day')['line_item_amount'].sum().sort_values(ascending = False)
by_day.plot(kind = 'bar');
# Write your findings
```

> 👉 **Write Your Findings Here**: Yes, best day is Monday, worst is Sunday.
-- - 


```python
# Are there better or worse days/times in terms of sales?
by_time = coffee_df.groupby('time_category')['line_item_amount'].sum().sort_values(ascending = False)
by_time.plot(kind = 'bar');
# Write your findings
```

> 👉 **Write Your Findings Here**: Yes, best time is Morning, worst is Night.
