#### __Variable:__
A variable is a labeled container that is used to store value(s). Look at it as a box where you label it with a name and store your item(s) in it.

**What is the purpose of a variable?**

It is to store, reference, and manipulate information.

**Naming Variables:**
* A variable name must start with a letter or the underscore character
* A variable name cannot start with a number
* A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
* Variable names are case-sensitive (number, Number, and NUMBER are three different variables)
* A variable name cannot be any of the Python keywords.

#### List methods

````python
# We have a list of fruits to buy
fruits = ["orange"]
# We need to add 'kiwi' to our list. Using append()
fruits.append('kiwi')

# fruits is now: ["orange", "kiwi"]
print(fruits)
````

````python
# Maybe we should get 'lemon' before 'kiwi' . Using insert()
fruits.insert(1,'lemon')

# fruits is now: ["orange", "lemon", "kiwi"]
print(fruits)
````

````python
# Remove the 'kiwi' from our list. Using pop()
fruits.pop()
# fruits is now: ["orange", "lemon"]
print(fruits)

# Remove 'orange' too using pop() pointing to the index of it
fruits.pop(0)

# fruits is now: ["lemon"]
print(fruits)
````

````python
# A. Access and print Arundhati Roy's first book
print(authors['Arundhati Roy']['books'][0])
````

````python
# B. We need to contact Brandon at work: access and print their work phone number
print(authors['Brandon Sanderson']['phone']['work'])
````

````python
# C. Create a list composed of all the genres available in our library
genres = []  # this creates empty list for you. You need to use append() to add items

genres.append(authors['Arundhati Roy']['genre'])
genres.append(authors['Brandon Sanderson']['genre'])
genres.append(authors['Kobo Abe']['genre'])

print(genres)
````

````python
# D. Create a list of all books available in our library
books = []  # this creates empty list for you. You need to use append() to add items

books.append(authors['Arundhati Roy']['books'])
books.append(authors['Brandon Sanderson']['books'])
books.append(authors['Kobo Abe']['books'])

print(books)
````

````python
# E. Arundhati has decided to expand their range of genres and tackle non-fiction
#    Modify their "genre" property to now be a list that also includes non-fiction
authors['Arundhati Roy']['genre'] = ['literary fiction', 'non-fiction']
````
